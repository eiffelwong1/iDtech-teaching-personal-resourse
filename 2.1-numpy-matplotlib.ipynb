```python
import random
import matplotlib.pyplot as plt
random.seed(0)
y = [i for i in range(20000)]
x = [random.random() for i in range(20000)]

plt.figure(figsize=(15,15))
plt.scatter(y, x)
plt.show()

plt.figure(figsize=(15,15))
plt.plot(y,x)
plt.show()
```

```python
x = [i for i in range(100)]
y = [x^3 + x^2 + x for x in range(100)]

plt.figure(figsize=(15,15))
plt.plot(y, x)
plt.show()
```
